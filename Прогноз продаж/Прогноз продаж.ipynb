````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from prophet import Prophet

import warnings
warnings.filterwarnings("ignore")

sns.set(style='darkgrid', palette='tab10', font_scale=1.2)
````

## Загрузка и анализ данных

Для выгрузки данных использовалась команда:
``` SQL
SELECT  date,
        SUM(quantity) AS quantities
FROM test.sales
WHERE item_id == '7d185936-7a60-11eb-ba7f-4a6a34607ded'
GROUP BY date
ORDER BY date
```

````python
data = pd.read_csv(r'D:\PC\Downloads\test.csv', index_col=[0], parse_dates=[0])
data = data.resample('d').sum()
````

````python
data.info()
data.sample(3)
````

````python
plt.figure(figsize=(15, 6))
plt.plot(data, alpha=0.7)
plt.plot(data.rolling(30).mean())
plt.plot(data.rolling(30).std())
plt.title('График продаж за 2021 год')
plt.ylabel('Кол-во продаж')
plt.xlabel('Дата')
plt.legend(['Продажи в сутки', 'Скользящее среднее за месяц',
           'Скользящее станд. отклонение за месяц'])

xdates = pd.date_range('2021-01', periods=12, freq='MS')
plt.xticks(ticks=xdates, labels=xdates.month)
plt.yticks(range(0, 90, 10))
````

Наблюдается снижение продаж в течение всего года с небольшим отскоком в ноябре.  
Взглянем на тренд и сезонность продаж:

````python
decomposed = seasonal_decompose(data)

plt.figure(figsize=(13, 8))
plt.subplot(311)
decomposed.trend.plot(ax=plt.gca())
plt.title('Тренд')

plt.subplot(312)
decomposed.seasonal.plot(ax=plt.gca())
plt.title('Сезонность')

plt.subplot(313)
decomposed.resid.plot(ax=plt.gca())
plt.title('Остатки')
plt.tight_layout()
````

Тренд повторяет график выше - снижение продаж в течение всего года.  
Сезонность стоит рассмотреть поближе.

````python
xdates = pd.date_range(start='2021-01-4', end='2021-01-11', freq='D')
decomposed.seasonal['2021-01-4': '2021-01-11'].plot()
plt.xticks(xdates, labels=xdates.dayofweek+1)
plt.title('Cезонность')
plt.xlabel('День недели')
plt.ylabel('Колебания продаж')
````

Колебания продаж повторяются каждые 7 дней. Сильные просадки в понедельник и пятницу, пики продаж приходятся на вторник и субботу.

## Предсказания

Для прогноза буду использовать Prophet.  
Для его работы нужно переименовать колонки датафрейма:

````python
data_prophet = data.reset_index()
data_prophet.columns = ['ds', 'y']
data_prophet.head()
````

````python
# Обучение модели и вывод прогноза
prophet = Prophet(yearly_seasonality=True,
                  weekly_seasonality=True).fit(data_prophet)

future = prophet.make_future_dataframe(periods=7, include_history=False)
preds = prophet.predict(future)
preds[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
````

````python
plt.plot(preds['ds'], preds['yhat'])
plt.fill_between(preds['ds'], preds['yhat_lower'],
                 preds['yhat_upper'], alpha=0.2)
plt.xlabel('Дата')
plt.ylabel('Продажи')
plt.title('Прогнозируемые продажи на следующие 7 дней')
plt.xticks(rotation=30, ha='right')
````

## Вывод
Профет прогнозирует постепенный рост продаж с начала января и имеет довольно серьезный разбег максимальных и минимальных возможных значений.  
Этот прогноз можно объяснить тем, что в январе прошлого года продажи так же показывали рост в первые недели. С увеличением объема данных точность должна улучшиться, поскольку на данный момент нам доступен только один год продаж и мы не можем с точностью определить являются ли подъемы и просадки продаж в определенные месяцы проявлением сезонности или случайными событями.
